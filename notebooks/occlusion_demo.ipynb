# Occlusion Mask Demo

This notebook visualizes a pitch/roll stabilized sensor, the scan frustum, and a 5-point body-attached occlusion profile.

The profile is authored with sensor-style `(azimuth, elevation, range)` points, then rotated with the aircraft body before occlusion is evaluated.


```python
import numpy as np

from occlusion_mask import OcclusionProfile, PlatformState, ScanVolume, make_visibility_figure

def sample_points():
    x_values = np.linspace(0.5, 8.0, 12)
    y_values = np.linspace(-3.0, 3.0, 9)
    z_values = np.linspace(-1.5, 2.5, 10)
    xx, yy, zz = np.meshgrid(x_values, y_values, z_values, indexing="ij")
    return np.column_stack([xx.ravel(), yy.ravel(), zz.ravel()])
```

```python
state = PlatformState.from_degrees(pitch_deg=-10.0, roll_deg=5.0, sensor_yaw_deg=0.0)
scan = ScanVolume.from_degrees(
    az_min_deg=-35.0,
    az_max_deg=35.0,
    el_min_deg=-25.0,
    el_max_deg=30.0,
    range_min_m=0.5,
    range_max_m=9.0,
)
profile = OcclusionProfile.from_sensor_az_el_range_degrees([
    (-40.0, 14.0, 3.0),
    (-20.0, 12.0, 2.8),
    (0.0, 8.0, 2.5),
    (20.0, 12.0, 2.8),
    (40.0, 14.0, 3.0),
])
```

```python
figure = make_visibility_figure(sample_points(), state, scan, profile)
figure
```

```python
for pitch_deg in (-15.0, 0.0, 15.0):
    state = PlatformState.from_degrees(pitch_deg=pitch_deg, roll_deg=0.0)
    fig = make_visibility_figure(sample_points(), state, scan, profile)
    fig.update_layout(title=f"Pitch sweep with sensor-space-authored boundary: {pitch_deg:+.1f} deg")
    fig.show()
```
